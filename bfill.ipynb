```python
import pandas as pd
```

```python
energy=pd.read_csv('ECdata.csv')
submit=pd.read_csv("PCprediction.csv")
```

```python
energy['Time'] = pd.to_datetime(energy['Time'])
submit['Time'] = pd.to_datetime(submit['Time'])
submit.drop(['w'], axis=1, inplace=True)
```

```python
combined_data = pd.concat([energy,submit], axis=0)
combined_data.head(10)
```

```python
# Sort the combined_data DataFrame by 'Time' and 'BS'
combined_data.sort_values(by=[ 'BS', 'Time'], inplace=True)
combined_data.head(10)
```

```python
combined_data.shape
```

```python
# Fill missing values using forward fill
#combined_data.fillna(method='ffill', inplace=True)
combined_data.fillna(method='bfill', inplace=True)

```

```python
merged_data = pd.merge(combined_data, submit, on=['Time', 'BS'], how='inner')
merged_data 
```

```python
merged_data.drop('Energy_y', axis=1, inplace=True)
```

```python
merged_data.rename(columns={'Energy_x': 'Energy'}, inplace=True)
merged_data
```

```python
submit= merged_data
# Create the 'ID' column by joining 'Time' and 'BS' columns
submit['ID'] = submit['Time'].astype(str) + '_' + submit['BS']

# Remove unnecessary columns and reorder columns
submit = submit[['ID', 'Energy']]

# Save the submission file
submit.to_csv('SampleSubmission__28_.csv', index=False)

submit.head()
```
